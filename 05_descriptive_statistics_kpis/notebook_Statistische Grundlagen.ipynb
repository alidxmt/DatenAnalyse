```python
# Verwendung von Bibliotheken zur Berechnung von Mittelwert und Median sowie deren Vergleich
import numpy as np
from scipy import stats

# Fall A: Normale Daten
normale_daten = [20, 22, 24, 25, 26, 28, 30]
print(f"Normaler Mittelwert: {np.mean(normale_daten):.2f}")

# Fall B: Ausreißer-Indikator (Der 'seltsame' Mittelwert)
ausreisser_daten = [20, 22, 24, 25, 26, 28, 1000]
print(f"Mittelwert mit Ausreißer: {np.mean(ausreisser_daten):.2f} (Nicht repräsentativ für die Mehrheit!)")

# Lösung: Getrimmter Mittelwert (entfernt die oberen/unteren 10%)
getrimmt = stats.trim_mean(ausreisser_daten, 0.1)
print(f"Lösung (Getrimmter Mittelwert): {getrimmt:.2f}")
```

## 2. Der Median (Robuste Zentralität)
**Best Practices:**
* **Resistenz gegen Schiefe:** Der Median ist der „mittlere“ Wert. Er ignoriert die Höhe der Extremwerte, was ihn zum Goldstandard für schiefe Daten macht (z. B. Gehälter, Immobilienpreise).
* **Ordinale Daten:** Verwende diesen für ranggeordnete Daten (z. B. „Auf einer Skala von 1-10...“).
* **Indikator:** Wenn `Mittelwert >> Median`, sind deine Daten stark rechtsschief.

```python
import pandas as pd

# Fall: Schiefe Verteilung
gehaelter = [45000, 48000, 50000, 52000, 55000, 60000, 1000000]
s = pd.Series(gehaelter)

print(f"Durchschnittsgehalt (Mean): ${s.mean():,.2f}")
print(f"Median Gehalt: ${s.median():,.2f}")

if s.mean() > s.median() * 1.5:
    print("Indikator: Signifikante Rechtsschiefe erkannt. Nutze den Median für die Berichterstattung.")
```

## 3. Der Modalwert (Kategorial & Bimodal)
**Best Practices:**
* **Nicht-numerische Daten:** Verwende diesen für Strings oder Kategorien, bei denen eine Mittelwertbildung unmöglich ist.
* **Indikator für „Versteckte Gruppen“:** Wenn deine Daten zwei Modalwerte haben (bimodal), deutet dies oft auf zwei verschiedene Populationen hin (z. B. Pendler-Hauptverkehrszeiten morgens vs. abends).
* **Rundung:** Bei kontinuierlichen Floats (Fließkommazahlen) sollten die Daten zuerst gerundet oder in Klassen („Bins“) eingeteilt werden, da sonst jeder Wert möglicherweise nur einmal vorkommt.

```python
# Fall: Bimodale Daten (Zwei Gipfel)
antworten = [1, 1, 1, 2, 5, 9, 10, 10, 10]
s_antworten = pd.Series(antworten)

modi = s_antworten.mode().tolist()
print(f"Erkannte Modalwerte (Modes): {modi}")

if len(modi) > 1:
    print("Indikator: Dies ist eine bimodale/multimodale Verteilung.")
```
