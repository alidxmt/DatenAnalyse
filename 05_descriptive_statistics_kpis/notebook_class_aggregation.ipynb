## Data Analysis Class

```python
class StockAnalyzer:
    def __init__(self, symbol, currency, description=""):
        self.symbol = symbol
        self.currency = currency
        self.data = None
        self.columns = ['DATE', 'OPEN', 'HIGH', 'LOW', 'CLOSE', 'TICKVOL', 'VOL', 'SPREAD']
        print("Accepted dataframe columns are:", self.columns)
    
    # Data Gathering Techniques (here from local file)
    def load_data_csv(self, path):
        if os.path.exists(path):
            self.data = pd.read_csv(path)
            print("Data loaded successfully.")
        else:
            print("File not found.")
    
    def load_api_data(self, api_url):
        # Placeholder for API data loading logic
        pass
    
    
    # Data Preparation
    
    # Data Quality Report 
    def report_quality(self):
        """Generates a summary of data issues without fixing them."""
        if self.data is None:
            return "No data loaded to report on."

        print(f"--- Data Quality Report for {self.symbol} ---")
        
        # 1. Missing Values
        missing = self.data.isnull().sum()
        print(f"\n[Missing Values]")
        print(missing[missing > 0] if missing.sum() > 0 else "None")

        # 2. Duplicates
        dup_count = self.data.duplicated(subset=['DATE']).sum()
        print(f"\n[Duplicates]\n- Found {dup_count} duplicate timestamps.")

        # 3. Logical Price Errors (The 'Physics' of the candle)
        # Check if High is actually the highest and Low is actually the lowest
        logical_errors = self.data[
            (self.data['HIGH'] < self.data['LOW']) |
            (self.data['HIGH'] < self.data['OPEN']) |
            (self.data['HIGH'] < self.data['CLOSE']) |
            (self.data['LOW'] > self.data['OPEN']) |
            (self.data['LOW'] > self.data['CLOSE'])
        ]
        print(f"\n[Price Logic Errors]\n- Found {len(logical_errors)} rows where High/Low logic is broken.")

        # 4. Negative Values (Spread/Volume)
        neg_spread = (self.data['SPREAD'] < 0).sum()
        neg_vol = (self.data['TICKVOL'] < 0).sum()
        print(f"\n[Negative Values]\n- Negative Spreads: {neg_spread}")
        print(f"- Negative Volume: {neg_vol}")

        # 5. Outliers (Simple 10% price jump check)
        # Calculates the % change between consecutive Close prices
        if 'CLOSE' in self.data.columns:
            pct_change = self.data['CLOSE'].pct_change().abs()
            extreme_jumps = (pct_change > 0.10).sum() # Jumps > 10%
            print(f"\n[Extreme Volatility]\n- Found {extreme_jumps} price jumps > 10% between rows.")

        print("\n" + "-"*40)
```
