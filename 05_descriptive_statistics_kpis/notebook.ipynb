## Task 2: The Random Day Test
Statistics is the art of being "less wrong" over time. To test our mindset, let’s pick a random day from our sales data. We’ll compare our Mean and Median against this specific day to see which one provides a better "real-world" guess. In the long run, which measure do you think will stay closer to the truth?

```python
def test_random_sale(data, mean, median):
    # Pick a random sale from the list
    random_sale = random.choice(data)
    
    dist_to_mean = abs(random_sale - mean)
    dist_to_median = abs(random_sale - median)
    
    print(f"--- Random Sale Picked: {random_sale} snacks ---")
    print(f"Distance to Mean ({mean:.2f}): {dist_to_mean:.2f}")
    print(f"Distance to Median ({median}): {dist_to_median}")
    
    if dist_to_median < dist_to_mean:
        print("RESULT: The Median was closer!")
    elif dist_to_mean < dist_to_median:
        print("RESULT: The Mean was closer!")
    else:
        print("RESULT: It's a tie!")
    print("-" * 40)

# Run the test 5 times to see the pattern
for i in range(5):
    test_random_sale(sales, mean_val, median_val)
```
