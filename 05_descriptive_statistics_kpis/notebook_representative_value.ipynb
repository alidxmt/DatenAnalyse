## Der Zufallstag-Test
Statistik ist die Kunst, im Laufe der Zeit "weniger falsch" zu liegen. Um unsere Denkweise zu testen, wählen wir einen zufälligen Tag aus unseren Verkaufsdaten aus. Wir vergleichen Mittelwert und Median mit diesem spezifischen Tag, um zu sehen, welcher Schätzwert in der "echten Welt" besser funktioniert. Welches Maß wird auf lange Sicht wohl näher an der Wahrheit bleiben?

```python
# So erhältst du einen Zufallswert. Jedes Mal, wenn wir das Programm ausführen, 
# erhalten wir eine neue Zahl und einen neuen Wert aus der Liste.
import random

# Erzeugt eine Zufallszahl zwischen 1 und 5
num = random.randint(1, 5)

liste_mit_zahlen = [21, 18, 32, 24, 45]

# Hinweis: In Python beginnt der Index bei 0. 
# Bei randint(1, 5) könnte der Index 5 einen Fehler verursachen, 
# wenn die Liste nur 5 Elemente hat.
print(num, liste_mit_zahlen[num-1])
```

```python
def test_random_sale(data, mean, median):
    # Wähle einen zufälligen Verkaufswert aus der Liste
    random_sale = random.choice(data)
    
    dist_to_mean = abs(random_sale - mean)
    dist_to_median = abs(random_sale - median)
    
    print(f"--- Zufälliger Tag gewählt: {random_sale} Snacks ---")
    print(f"Abstand zum Mittelwert ({mean:.2f}): {dist_to_mean:.2f}")
    print(f"Abstand zum Median ({median}): {dist_to_median}")
    
    if dist_to_median < dist_to_mean:
        print("ERGEBNIS: Der Median war näher dran!")
    elif dist_to_mean < dist_to_median:
        print("ERGEBNIS: Der Mittelwert war näher dran!")
    else:
        print("ERGEBNIS: Unentschieden!")
    print("-" * 40)

# Führe den Test 5 Mal aus, um ein Muster zu erkennen
for i in range(5):
    test_random_sale(sales, mean_val, median_val)
```
