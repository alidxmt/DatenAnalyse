### **Objective**
In this notebook, you will learn how to:
1. **Audit** a dataset for missing values and duplicates.
2. **Clean** data using imputation (filling gaps) and removal.
3. **Analyze** the central tendency (Mean, Median, Mode) of your variables.

### Step 1: Initialize Environment
**Instructions:**
* Import the `pandas` library.
* Load the `titanic.csv` file into a DataFrame.
* Preview the first few rows to understand the structure.

```python
import pandas as pd

# Loading dataset
df = pd.read_csv('titanic.csv')
df.head()
```

### Step 2: Data Quality Audit
**Instructions:**
* Use `.isnull().sum()` to count missing entries per column.
* Use `.duplicated().sum()` to identify identical rows.

```python
print("--- Missing Values Check ---")
print(df.isnull().sum())

print("\n--- Duplicate Rows Check ---")
print(f"Total duplicates found: {df.duplicated().sum()}")
```

### Step 3: Data Cleaning Strategy
**Instructions:**
* **Drop Duplicates:** Remove redundant rows.
* **Impute Age:** Fill missing age values with the **Median**.
* **Impute Embarked:** Fill missing ports with the **Mode**.

```python
# 1. Remove duplicates
df = df.drop_duplicates()

# 2. Handle missing data
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

print("Cleaning complete. Remaining nulls:")
print(df[['age', 'embarked']].isnull().sum())
```

### Step 4: Final Statistics
**Instructions:**
* Calculate **Mean** and **Median** for continuous numerical data.
* Identify the **Mode** for categorical data.

```python
num_cols = ['age', 'fare', 'sibsp', 'parch']
cat_cols = ['class', 'embarked', 'gender', 'survived', 'country']

print("SUMMARY: NUMERICAL DATA")
summary_stats = pd.DataFrame({
    'Mean': df[num_cols].mean(),
    'Median': df[num_cols].median()
})
display(summary_stats)

print("\nSUMMARY: CATEGORICAL DATA (MODE)")
for col in cat_cols:
    print(f"{col.title():<10}: {df[col].mode()[0]}")
```

### 🌟 Step 5: The Challenge Task
**Instructions:**
* Use a **filter** to calculate the survival rate (mean of `survived`) for **Females** vs **Males**.
* Identify which **Country** had the highest number of survivors.
* *Hint: Use `df[df['gender'] == 'female']['survived'].mean()`*

```python
# WRITE YOUR CHALLENGE CODE HERE

female_survival = df[df['gender'] == 'female']['survived'].mean()
male_survival = df[df['gender'] == 'male']['survived'].mean()

print(f"Female Survival Rate: {female_survival:.2%}")
print(f"Male Survival Rate: {male_survival:.2%}")

top_country = df[df['survived'] == 1]['country'].value_index()[0]
print(f"Country with most survivors: {top_country}")
```
