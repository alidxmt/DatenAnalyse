### **Lernziele**
In diesem Notebook lernst du:
1. **Datenprüfung:** Identifizierung von fehlenden Werten (Missing Values) und Duplikaten.
2. **Datenbereinigung:** Umgang mit Lücken durch Imputation (Auffüllen) oder Löschen.
3. **Statistische Analyse:** Berechnung von Lagemaßen (Mittelwert, Median, Modus).

### Schritt 1: Umgebung vorbereiten
**Anweisungen:**
* Importiere die Bibliothek `pandas`.
* Lade die Datei `titanic.csv` in einen DataFrame.
* Nutze `.head()`, um die ersten Zeilen und die Struktur zu prüfen.

```python
import pandas as pd

# Datensatz laden
df = pd.read_csv('titanic.csv')
df.head()
```

### Schritt 2: Prüfung der Datenqualität
**Anweisungen:**
* Nutze `.isnull().sum()`, um fehlende Werte pro Spalte zu zählen.
* Nutze `.duplicated().sum()`, um identische Zeilen zu finden, die die Analyse verfälschen könnten.

```python
print("--- Prüfung auf fehlende Werte ---")
print(df.isnull().sum())

print("\n--- Prüfung auf Duplikate ---")
print(f"Anzahl der Duplikate: {df.duplicated().sum()}")
```

### Schritt 3: Strategie zur Datenbereinigung
**Anweisungen:**
* **Duplikate entfernen:** Lösche redundante Zeilen.
* **Alter (Age) auffüllen:** Ersetze fehlende Werte durch den **Median** (robuster gegenüber Ausreißern).
* **Zustiegshafen (Embarked) auffüllen:** Ersetze Lücken durch den **Modus** (den häufigsten Wert).

```python
# 1. Duplikate entfernen
df = df.drop_duplicates()

# 2. Fehlende Werte bearbeiten
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

print("Bereinigung abgeschlossen. Verbleibende Nullwerte:")
print(df[['age', 'embarked']].isnull().sum())
```

### Schritt 4: Deskriptive Statistik
**Anweisungen:**
* Berechne den **Mittelwert (Mean)** und **Median** für numerische Daten.
* Bestimme den **Modus (Mode)** für kategorische Spalten (Geschlecht, Klasse, etc.).

```python
num_cols = ['age', 'fare', 'sibsp', 'parch']
cat_cols = ['class', 'embarked', 'gender', 'survived', 'country']

print("ZUSAMMENFASSUNG: NUMERISCHE DATEN")
summary_stats = pd.DataFrame({
    'Mittelwert': df[num_cols].mean(),
    'Median': df[num_cols].median()
})
display(summary_stats)

print("\nZUSAMMENFASSUNG: KATEGORISCHE DATEN (MODUS)")
for col in cat_cols:
    print(f"{col.title():<10}: {df[col].mode()[0]}")
```

### 🌟 Schritt 5: Die Challenge-Aufgabe
**Anweisungen:**
* Filtere die Daten und berechne die **Überlebensrate** (Mittelwert von `survived`) für **Frauen (female)** im Vergleich zu **Männern (male)**.
* Finde heraus, welches **Land (country)** die höchste Anzahl an Überlebenden hatte.
* *Hinweis: Nutze `df[df['gender'] == 'female']['survived'].mean()`*

```python
# HIER DEINEN CODE FÜR DIE CHALLENGE EINGEBEN

female_survival = df[df['gender'] == 'female']['survived'].mean()
male_survival = df[df['gender'] == 'male']['survived'].mean()

print(f"Überlebensrate Frauen: {female_survival:.2%}")
print(f"Überlebensrate Männer: {male_survival:.2%}")

top_country = df[df['survived'] == 1]['country'].value_counts().idxmax()
print(f"Land mit den meisten Überlebenden: {top_country}")
```
