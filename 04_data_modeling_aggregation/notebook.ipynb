# Data Modelling

```python
import pandas as pd

# === Step 1: Define Entities and Attributes ===
entities = {
    "Trader": ["TraderID (PK)", "Name", "Email"],
    "TradingAccount": ["AccountID (PK)", "TraderID (FK)", "AccountType"],
    "Stock": ["StockID (PK)", "Symbol", "CompanyName", "Sector"],
    "Trade": ["TradeID (PK)", "TraderID (FK)", "TradeTime"],
    "TradeDetails": ["TradeID (FK)", "StockID (FK)", "Quantity"]
}

# Display entities with their attributes
for entity, attrs in entities.items():
    print(f"Entity: {entity}")
    for attr in attrs:
        print(f"  - {attr}")
    print()

# === Step 2: Define Relationships ===
relationships = [
    {"from": "Trader", "to": "TradingAccount", "type": "1:1"},
    {"from": "Trader", "to": "Trade", "type": "1:N"},
    {"from": "Trade", "to": "TradeDetails", "type": "1:N"},
    {"from": "Stock", "to": "TradeDetails", "type": "1:N"}
]

# Display relationships as a table
rel_df = pd.DataFrame(relationships)
rel_df
```
