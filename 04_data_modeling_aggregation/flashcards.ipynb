```python
import pandas as pd
import random
import ipywidgets as widgets
from IPython.display import display, HTML

# Load CSV
df = pd.read_csv("flashcards.csv")

# Shuffle cards
df = df.sample(frac=1).reset_index(drop=True)

current_index = 0
showing_answer = False

# Card styling (square card)
card = widgets.HTML(
    value="",
    layout=widgets.Layout(
        width="400px",
        height="400px",
        border="2px solid black",
        display="flex",
        align_items="center",
        justify_content="center",
        padding="20px"
    )
)

button_show = widgets.Button(description="Show Answer")
button_next = widgets.Button(description="Next Card")

def update_card():
    global showing_answer
    if showing_answer:
        text = df.loc[current_index, "Answer"]
    else:
        text = df.loc[current_index, "Question"]
        
    card.value = f"""
    <div style="
        width:100%;
        height:100%;
        display:flex;
        align-items:center;
        justify-content:center;
        text-align:center;
        font-size:20px;
        font-family:Arial;
    ">
        {text}
    </div>
    """

def on_show_clicked(b):
    global showing_answer
    showing_answer = not showing_answer
    update_card()

def on_next_clicked(b):
    global current_index, showing_answer
    current_index = random.randint(0, len(df)-1)
    showing_answer = False
    update_card()

button_show.on_click(on_show_clicked)
button_next.on_click(on_next_clicked)

update_card()

display(card, widgets.HBox([button_show, button_next]))
```
